```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/data.csv')
```

```python
df.head()
```

```python
df['time'] = pd.Series(range(1, len(df)+1))
```

```python
df['class'].describe()
```

```python
dict = {'neutral' : 0, 'back' : 1, 'forward' : 2, 'left': 3, 'right' : 4}
df['class'].replace(dict, inplace=True)
```

```python
df.head()
```

```python
X = df.drop(['class', 'time'], axis=1)
```

```python
X
```

```python
y = df['class']
```

```python
y
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier(random_state=17, n_estimators=200)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
forest.fit(X_train, y_train)
```

```python
forest.score(X_train, y_train)
```

```python
forest.feature_importances_
```

```python
pred = forest.predict(X_test)
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(pred, y_test, average='weighted')
```
